```python
def outlier_handler(col):
    d = x[col]
    q1 = d.quantile(0.25)
    q3 = d.quantile(0.75)
    iqr = q3-q1
    lower = q1 - 1.5*iqr
    upper = q3 + 1.5*iqr
    
    d = d.clip(lower,upper)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data["Sex"] = encoder.fit_transform(data["Sex"])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import AdaBoostRegressor, RandomForestRegressor
from sklearn.linear_model import HuberRegressor, RANSACRegressor, LinearRegression, Ridge, Lasso
from xgboost import XGBRegressor

model = KNeighborsRegressor(n_neighbors=5, weights="distance", algorithm="brute", p=3)
model.fit(x_train, y_train)
ans = model.predict(x_test)
```

```python
ans = pd.DataFrame(ans, columns=["predicted"])
```
